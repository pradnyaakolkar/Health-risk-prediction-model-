```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np

# Load models
heart_model = joblib.load("heart_disease_model.pkl")
diabetes_model = joblib.load("diabetes_model.pkl")

st.title("Health Risk Prediction App")
st.subheader("Predict Heart Disease or Diabetes")

option = st.selectbox("Choose Prediction Type", ("Heart Disease", "Diabetes"))

if option == "Heart Disease":
    st.markdown("### Enter the following details:")
    age = st.number_input("Age", min_value=1)
    sex = st.selectbox("Sex (0 = Female, 1 = Male)", [0, 1])
    cp = st.slider("Chest Pain Type (0-3)", 0, 3)
    trestbps = st.number_input("Resting Blood Pressure")
    chol = st.number_input("Cholesterol")
    fbs = st.selectbox("Fasting Blood Sugar > 120 mg/dl (1 = True, 0 = False)", [0, 1])
    restecg = st.slider("Rest ECG (0-2)", 0, 2)
    thalach = st.number_input("Max Heart Rate Achieved")
    exang = st.selectbox("Exercise Induced Angina (1 = Yes, 0 = No)", [0, 1])
    oldpeak = st.number_input("ST depression induced by exercise")
    slope = st.slider("Slope of the peak exercise ST segment (0-2)", 0, 2)
    ca = st.slider("Number of major vessels colored by fluoroscopy (0-3)", 0, 3)
    thal = st.slider("Thalassemia (1 = Normal; 2 = Fixed Defect; 3 = Reversible Defect)", 0, 3)

    if st.button("Predict Heart Disease"):
        input_data = np.array([[age, sex, cp, trestbps, chol, fbs, restecg,
                                thalach, exang, oldpeak, slope, ca, thal]])
        prediction = heart_model.predict(input_data)
        st.success("Prediction: " + ("Positive (Heart Disease)" if prediction[0] == 1 else "Negative"))

elif option == "Diabetes":
    st.markdown("### Enter the following details:")
    pregnancies = st.number_input("Pregnancies", min_value=0)
    glucose = st.number_input("Glucose")
    blood_pressure = st.number_input("Blood Pressure")
    skin_thickness = st.number_input("Skin Thickness")
    insulin = st.number_input("Insulin")
    bmi = st.number_input("BMI")
    dpf = st.number_input("Diabetes Pedigree Function")
    age = st.number_input("Age", min_value=1)

    if st.button("Predict Diabetes"):
        input_data = np.array([[pregnancies, glucose, blood_pressure, skin_thickness,
                                insulin, bmi, dpf, age]])
        prediction = diabetes_model.predict(input_data)
        st.success("Prediction: " + ("Positive (Diabetes)" if prediction[0] == 1 else "Negative"))
```
